# Eigenverdier

## Oppgave

- Er $2$ en egenverdi til matrisen
$$A = 
\begin{bmatrix}
  3  & -2  &  0  &  0 \\
  1  &  0  &  0  &  0 \\
  1  & -2  &  2  &  1 \\
  0  &  2  & -2  &  5 
\end{bmatrix}
$$

La $P$ være matrisen
$$
P 
=
\begin{bmatrix}
  1  &  2  &  0  &  0 \\
  1  &  1  &  0  &  0 \\
  1  &  1  &  1  &  1 \\
  0  &  0  &  1  &  2 
\end{bmatrix}
$$

- Regn ut matriseproduktet
$$AP$$

- Finn tall $\lambda_1, \lambda_2, \lambda_3$ og $\lambda_4$ slik at 
$$
AP
= 
P
\begin{bmatrix}
  \lambda_1  &  0  &  0  &  0 \\
  0  &  \lambda_2  &  0  &  0 \\
  0  &  0  &  \lambda_3  &  0 \\
  0  &  0  &  0  &  \lambda_4 
\end{bmatrix}
$$
