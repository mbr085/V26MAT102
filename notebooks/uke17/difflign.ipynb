# System av differensialligninger

## Oppgave

Finn funksjoner $x(t)$ og $y(t)$ slik at 
$$
\begin{eqnarray}
x'(t) &=& 4x(t) - 3y(t)\\
y'(t) &=& 2x(t) - y(t)
\end{eqnarray}
$$
med startbetingelsen $x(0) = 4$ og $y(0) = 3$.

## Løsningsstrategi

Gitt et system av differensialligninger på formen
$$
\begin{eqnarray}
x'(t) &=& px(t) - ry(t)\\
y'(t) &=& qx(t) - sy(t)
\end{eqnarray}
$$
med startbetingelsen $x(0) = e$ og $y(0) = f$.

1. Finn egenverdier $\lambda_1$ og $\lambda_2$ for matrisen $\begin{bmatrix} p & q \\ r & s\end{bmatrix}$ og tilhørende egenvektorer $\vec w_1$ og $\vec w_2$.
2. Skriv vektoren $\vec v_0 = \begin{bmatrix} e \\ f \end{bmatrix}$ på formen $\vec v_0 = a \vec w_1 + b \vec w_2$.

Da er 
$$v(t) = ae^{\lambda_1 t}\vec w_1 + be^{\lambda_2 t} \vec w_2$$
en løsning til differensialligningssystemet
$$
\vec v'(t) = 
\begin{bmatrix} p & q \\ r & s\end{bmatrix} v(t)
$$
med $\vec v(0)= \vec v_0$

Funksjonene $x(t)$ og $y(t)$ er gitt ved komponentene til vektoren $\vec v(t)$.

## Konkret oppgave

1. Finn egenverdier $\lambda_1$ og $\lambda_2$ for matrisen $\begin{bmatrix} 4 & -3 \\ 2 & -1\end{bmatrix}$ og tilhørende egenvektorer $\vec w_1$ og $\vec w_2$.
2. Skriv vektoren $\vec v_0 = \begin{bmatrix} 4 \\ 3 \end{bmatrix}$ på formen $\vec v_0 = a \vec w_1 + b \vec w_2$.
3. Finn funksjoner $x(t)$ og $y(t)$ slik at 
$$
\begin{eqnarray}
x'(t) &=& 4x(t) - 3y(t)\\
y'(t) &=& 2x(t) - y(t)
\end{eqnarray}
$$
med startbetingelsen $x(0) = 4$ og $y(0) = 3$.
