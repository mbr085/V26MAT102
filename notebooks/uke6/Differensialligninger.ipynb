# Ordinære differensialligninger

En differensialligning er en ligning som inneholder deriverte funksjoner.

Se også kapittel 4 i [kompendiet](https://mitt.uib.no/courses/45603/files?preview=5619281) for mere om dette.

```python
# @title Import av moduler
import numpy as np
import matplotlib.pyplot as plt
```

Vi ser på **førsteordens differensialligninger** på formen
$$x'(t) = f(t,x(t))$$
for en funksjon
$$
\begin{eqnarray}
\mathbb{R}^2 &\xrightarrow f& \mathbb{R}, \newline
(t, x) &\mapsto& f(t, x).
\end{eqnarray}
$$

## Tenkt eksempel

Vi har en mær med laks. Ukeldigvis er det også lakselus i mæren med en vekt $x(t)$ som endrer seg med tiden $t$. Endringsraten $f(t, x(t))$ for vekten til lakselusene avhenger av vekten $x(t)$ til lakselusen og tiden $t$ på en måte som kanskje ikke er velforstått. Grunnen til at endringsraten variterer med tiden kan være faktorer som foringssyklus og temperatur. Uten å nærmere informasjon om disse faktorer kan vi prøve å finne ut hvordan endringsraten endrer seg med tiden for å lage en modell for hvordan vekten til lakselusene endrer seg med tiden under forskjellige forhold.

## Eksempel med tall

La oss se på tilfellet der $f(t,x) = 2$ for alle $x$ og $t$
slik at vi ser på differensialligningen
$$x'(t) = 2.$$

Fra videregående skole har vi lært at den deriverte til funksjonen $x_0(t) = 2t$ er $x_0'(t) = 2$. 

Dette er altå en løsning $x_0(t)$ til differensialligningen.

Dersom $x(t)$ er en funksjon med $x'(t) = 2$ for all $t$, da er $x'(t) - x_0'(t) = 0$ for alle $t$.

Lar vi 
$$g(t) = x(t) - x_0(t)$$ blir 
$$g'(t) = x'(t) - x_0'(t) = 0$$
for alle $t$. 

Derfor er funksjonen $g(t)$ konstant, så
$$g(t) = x(t) - x_0(t) = C$$ 
for en konstant $C$. 

Det vil si
$$x(t) - x_0(t) = x(t) - 2t= C,$$
så
$$x(t) = 2t + C.$$

Vi kan konkludere at enhver løsning til $x'(t) = 2$ er på formen $x(t) = x_C(t) = 2t + C$ for en konstant $C$.

```python
# @title Plotting av løsningskurver
t = np.linspace(-2,2,50)
for i in range(-2,3):
    plt.plot(t, 2*t + i, label=f'C = {i}')
plt.legend()
plt.title(r"Løsninger til x′(t)=2");
```

Hvis vi har
$$x'(t) = 2, \qquad x(0) = 0$$
da er det bare løsningen $x_0(t) = 2t$ som oppfylder andre betingelse, altså $x_0(0) = 0$.

## Eksempel med tall

Løs differensialligningen
$$x'(t) = x(t), \quad \text{med} \quad x(0) = 2$$

Her kjenner vi en løsning til $x'(t) = x(t)$, nemlig $x_1(t) = e^t$. 

Bemerk at $x_1(0) = 1$, så $x_1(x)$ er ikke løsningen vi er på jakt etter. 

Hvis $x(t)$ er en annen løsning, da gir regelen for derivasjon av brøker at
$$ \left(\frac{x(t)}{x_1(t)}\right)' = 0$$

### Oppgave:

**Sjekk påstanden over.**

Bruk gjerne regelen
$$\left(\frac{f}{g}\right)' = \frac{f' \cdot g - f\cdot g'}{g^2}$$


Det vil si at funksjoen $\frac{x(t)}{x_1(t)}$ er konstant, så $x(t) = C x_1(t) = Ce^t$ for en konstant $C$. 

Skriver vi $x_C(t) = C e^t$, så er $x_C(0) = C$, Derfor er $x_2(t) = 2e^t$
en funksjon som oppfylder de gitte betingelsene:
$$x_2'(t) = x_2(t), \quad \text{med} \quad x_2(0) = 2.$$

# Hastighetsvektor

La oss tenke på $x(t)$ som vår høyde over sjøen når vi har beveget oss lengden $t$ langs en vei.

Da vil $x'(t)$ være stigningstallet til veien, det vil si hvor bratt veien er akkuart etter vi har kjørt lengden $t$.

Kjører bilen med konstant horisontal fart en meter per sekund kan posisjonen til bilen kan skrives 
$$(t, x(t)).$$

Deriverer vi i begge koordinatene får vi hastigheten til bilen:
$$(1, x'(t)).$$

```python
# @title Plotting av graf med pil
# import matplotlib.pyplot as plt
import matplotlib.patches as patches
# import numpy as np

# Definer funksjonen x(t) = sin(t)
def x(t):
  return np.sin(t)

# Definer funksjonen f(t, x) = x
def f(t, x):
  return np.cos(t)

t_start = 0
t_end = t_start + 2*np.pi

# Startpunkt for pilen
pil_t_start = 0.2
pil_x_start = x(pil_t_start)

# Sluttpunkt for pilen
pil_t_end = pil_t_start + 1
pil_x_end = pil_x_start + f(pil_t_start, pil_x_start)

t = np.linspace(t_start, t_end, 100)


# Lag en FancyArrowPatch
plt.figure(figsize=(8,6))
arrow = patches.FancyArrowPatch((pil_t_start, pil_x_start), (pil_t_end, pil_x_end),
                                 mutation_scale=20,
                                 #arrowstyle="->",
                                 color="k")

# Legg til pilen til plottet
ax = plt.gca()
ax.add_patch(arrow)
ax.plot(t, x(t))
ax.plot(pil_t_start, pil_x_start, 'o')

# Sett grenser for aksene for å vise pilen
# ax.set_xlim([0, 3.2])
# ax.set_ylim([0, 40])

# Sett akser og tittel
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title("Hastighetsvektor (1,x′(t)) for en bil på en fjellvei.");
```

# Hastighetsvektorer for løsning til differensialligning

Førsteordens differensialligninger på formen
$$x'(t) = f(t,x(t))$$
kan tolkes **grafisk** ved hjelp av såkalte **retningsfelt**. Retningsfelt består av hastighetsvektorer for mange forskjellige løsninger til differensialligningen. 

La $x(t)$ være en løsning til diffligningen over, og se på **parametriseringen** $t \mapsto (t, x(t))$ for **grafen** til funksjonen $x(t)$. 

Tænker vi på $(t, x(t))$ som posisjonen til en bil som beveger seg med konstant horisontal fart en meter per sekund langs en vei i et kuppert landskap,
da er hastighetsvektoren til tidspunktet $t$ git ved 
$$(1,x'(t)) = (1, f(t,x(t))).$$
Siden patikkelen til tiden $t$ befinner seg i posisjon $(t, x(t))$ kan vi grafisk angi hastighetsvektoren ved å tegne en pil fra punktet i $\mathbb{R}^2$ med koordinater $(t, x(t))$
til punktet med koordinater $(t, x(t)) + (1, f(t,x(t))) = (t+1, x(t) + f(t, x(t)))$.

### Oppgave:

**La $a > 0$, og se på parametriseringen $t \mapsto (at, x(at))$ for grafen til funksjonen $x(t)$. Hva er den horisontale farten? Hvordan kan vi grafisk angi hastighetsvektoren for en bil som beveger seg på denne måten?**

### Eksempel

La $f(t,x) = x$ slik at vi får diffligningen 
$$x'(t) = x(t).$$ 

Vi ser på løsningen $x(t) = 2e^t$ som vi fant over. 

For $t = 2$ er 
$$f(t, x(t)) = x(t) = x(2) = 2 e^2.$$ 

Vi vil altså tegne en pil som begynner i punktet 
$$(t, x(t)) = (2, x(2)) = (2, 2e^2)$$ 
og som ender i punktet 
$$(t+1, x(t) + f(t, x(t))) = (1 + 2, 2e^2 + 2e^2) = (3, 4e^2).$$ 

Se figuren under. Koden kan du ignorere.

```python
# @title Plotting av graf med pil
# import matplotlib.pyplot as plt
import matplotlib.patches as patches
# import numpy as np

def x(t):
  return 2 * np.exp(t)

# Definer funksjonen f(t, x) = x
def f(t, x):
  return x

t_start = 0
t_end = 3

# Startpunkt for pilen
pil_t_start = 2
pil_x_start = x(pil_t_start)

# Sluttpunkt for pilen
pil_t_end = pil_t_start + 1
pil_x_end = pil_x_start + f(pil_t_start, pil_x_start)

t = np.linspace(t_start, t_end, 100)


# Lag en FancyArrowPatch
plt.figure(figsize=(8,6))
arrow = patches.FancyArrowPatch((pil_t_start, pil_x_start), (pil_t_end, pil_x_end),
                                 mutation_scale=20,
                                 #arrowstyle="->",
                                 color="k")

# Legg til pilen til plottet
ax = plt.gca()
ax.add_patch(arrow)
ax.plot(t, x(t))
ax.plot(pil_t_start, pil_x_start, 'o')

# Sett grenser for aksene for å vise pilen
# ax.set_xlim([0, 3.2])
# ax.set_ylim([0, 40])

# Sett akser og tittel
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title('Grafen til $x(t)$ og pil fra $(2, 2e^2)$ til $(3, 4e^2)$');
```

## Retningsfelt

Hvis vi ikke kjenner noen løsning til vår differensialligning kan vi bruke
$$x'(t) = f(t, x(t))$$
til å tegne retningsvektorer $f(t, x)$ i punkter $(t, x)$. 

Dette er illustrert under for differensialligningen
$$x'(t) = f(t, x(t)) = x(t).$$

```python
# @title På vei mot Eulers metode
import numpy as np
import matplotlib.pyplot as plt

# Definer funksjonen x_2(t) = 2e^t
def x_2(t):
  return 2 * np.exp(t)

# Definer funksjonen f(t, x) = x
def f(t, x):
  return x

n = 20


# Definerer x- og y-verdiene
t = np.linspace(0, 3, num=n+1)
x = np.linspace(0, 40, num=n+1)

# Lager et firkantet gitter av de oppgitte x- og y-verdiene
T, X = np.meshgrid(t, x)

# Definerer f med hensyn på T og X
F = f(T, X)

# Definerer "steget" i t- og x-retning
dt = np.ones_like(T)
dx = F

# plt.quiver tar for seg å lage alle tangentpilene [dt, dx] i alle punktene som opptrer i gitteret vi lagde.
plt.quiver(T, X, dt, dx, scale=500, scale_units='width', color='black', angles='xy')
plt.plot(t, x_2(t), label='x(t) = 2e^t')
plt.ylim((0, 50))
plt.grid()
plt.show()
```
