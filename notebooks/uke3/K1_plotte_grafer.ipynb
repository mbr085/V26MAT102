## Plotting av grafer

Det er mulig å tegne mange forskjellige figurer med pakken `matplotlib.pyplot`. 

For å bruke denne pakken må vi først importere den. Grunnen til dette er at det 
finnes mange pakker, og python ville blit alt for klossete og treg hvis alle verdens pakker ble inkludert som standard. 

Det finnes mange måter å importere pakker på. Vi kan velge å gi pakker et kort navn som vi enkelt kan huske. For pakken `matplotlib.pyplot` er det 
vanlig å gi den det korte navnet `plt`. Vi importerer også pakken `numpy` det korte navnet `np` 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Pakken `numpy` som vi nettopp har importert med det korte navnet `np` kan gi oss tall som ligger jevnt mellom to endeverdier:

```python
x = np.linspace(-2*np.pi, 2*np.pi, num=10)
```

Hvis vi vil se tallene vi har generert kan vi enkelt og greit spørre jupyter hva `x` er for noe:

```python
x
```

Pakken `numpy` (forkortet `np`) har mange innebyggede funksjoner som kan anvendes på lister av tall som listen `x` over. 

For eksempel sinus funksjonen. Vi anvender `np.sin` funksjonen 
på listen `x` og gi resultatet navnet `y`:

```python
y = np.sin(x)
```

Vi kan nå plotte `x`-verdiene mot verdiene til `y = np.sin(x)`:

```python
plt.plot(x, y)
```

Denne figuren er litt rar. Den består av linjestykker mellom de ti punktene i planet med koordinater $(x_0, y_0), (x_1, y_1), \dots, (x_9, y_9)$ listet opp under.

```python
list(zip(x, y))
```

### Jevnere figur
Hvis vi velger å ta med flere `x`-verdier og `y`-verdier får vi en figur som viser bedre at det er sinusfunksjonen vi jobber med:

```python
xx = np.linspace(-2*np.pi, 2*np.pi, num=100)
yy = np.sin(xx)
```

```python
plt.plot(xx, yy);
```

Når vi tegner en figur kan det være hjelpsomt å ha med litt tekst i figuren, som tittel og betegnelse på aksene. Dette går greit med matplotlib:

```python
plt.plot(xx, np.sin(xx))
plt.title('Grafen til sinus funksjonen')
plt.xlabel('x')
plt.ylabel('sin(x)');
```

Det går også an å tegne koordinatakser. Denne typen arbeid er KI veldig god til å hjelpe med.

```python
ax = plt.gca()
ax.plot(xx, np.sin(xx))
#plt.title('Grafen til sinus funksjonen')
#ax.set_xlabel('x')
ax.set_xlabel('x', size=14, labelpad=-24, x=1.03)
ax.set_ylabel('sin(x)', size=14, labelpad=-21, y=1.02, rotation=0)
#ax.set_ylabel('sin(x)')
# Set bottom and left spines as x and y axes of coordinate system
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
