# Determinanter (MIP 10.6)

Vi sier at en matrise $A$ er **kvadratisk** hvis den har like mange rader og kolonner.

Determinanten $|A|$ til en kvadratisk matrise er et tall.

For en $1 \times 1$-matrise $A = [a]$ er $|A| = a$.

For en $2 \times 2$-matrise
$$
A = 
\left[\begin{array}{cc}
a & b\\
c & d
\end{array}\right]
$$
er
$$
|A| = 
\left|\begin{array}{cc}
a & b\\
c & d
\end{array}\right|
= ad - bc
$$


## Eksempel

For
$$
A = 
\left[\begin{array}{cc}
1 & 2\\
3 & 4
\end{array}\right]
$$
er
$$
|A| = 
\left|\begin{array}{cc}
1 & 2\\
3 & 4
\end{array}\right|
= 1 \cdot 4 - 2 \cdot 3 = 4 - 6 = -2
$$

---

For en $3 \times 3$-matrise
$$
A = 
\left[\begin{array}{ccc}
a & b & c\\
d & e & f \\
g & h & i
\end{array}\right]
$$
er
$$
|A| = 
\left|\begin{array}{ccc}
a & b & c\\
d & e & f \\
g & h & i
\end{array}\right|
= 
a\left|\begin{array}{cc}
e & f\\
h & i
\end{array}\right| -
b\left|\begin{array}{cc}
d & f\\
g & i
\end{array}\right| +
c\left|\begin{array}{cc}
d & e\\
g & h
\end{array}\right|
$$


## Eksempel

For 
$$
I = 
\left[\begin{array}{ccc}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right]
$$
er
$$
|I| = 
\left|\begin{array}{ccc}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right|
= 
1\left|\begin{array}{cc}
1 & 0\\
0 & 1
\end{array}\right| -
0\left|\begin{array}{cc}
0 & 0\\
0 & 1
\end{array}\right| +
0\left|\begin{array}{cc}
0 & 1\\
0 & 0
\end{array}\right|
= 1 -0 + 0 = 1
$$

For 
$$
A = 
\left[\begin{array}{ccc}

\end{array}\right]
$$
er
$$
|A| = 
\left|\begin{array}{ccc}

## Generell formel for determinant

Gitt en kvadratisk matrise
$$A = 
\begin{bmatrix}
  a_{00} & a_{01} & \dots & a_{0n} \\
  a_{10} & a_{11} & \dots & a_{1n} \\
  \vdots & \vdots & \ddots & \vdots \\
  a_{n0} & a_{n1} & \dots & a_{nn} 
\end{bmatrix}
$$
og en $i$ mellom $0$ og $n$, skriv $A_i$ for matrisen oppnådd ved å
fjene rad $0$ og kolonne $i$ fra matrisen $A$.

Da er
$$|A| = a_{00}|A_0| - a_{01}|A_1| + \dots + (-1)^n a_{0n}|A_n|$$

Dette er et eksempel på en såkallt *rekursiv* definisjon. For å regne ut determinanten til
en $n \times n$ matrise må vi først regne ut determinanten til en drøss med $(n-1) \times (n-1)$ matriser,
og for å regne disse ut må vi regne ut determinanter til $(n-2) \times (n-2)$ matriser. Følger vi denne tråden ender vi opp
med å regne ut matriser til en skog med $1 \times 1$ matriser, som vi vet hvordan vi gjør.

## Teorem
**Antall løsninger av ligningssystemer**

Systemet
$$
\begin{array}{ccccccccc}
a_{00} x_0 &+& a_{01} x_1& + & \dots & + & a_{0n}x_n & = & b_0 \\
a_{10} x_0 &+& a_{11} x_1& + & \dots & + & a_{1n}x_n & = & b_1 \\
\vdots&&\vdots&&&&\vdots&&\vdots \\
a_{n0} x_0 &+& a_{n1} x_1& + & \dots & + & a_{nn}x_n & = & b_n 
\end{array}
$$
har en entydig løsning hvis og bare hvis
$$
\left|
\begin{array}{cccc}
a_{00} & a_{01} & \dots & a_{0n} \\
a_{10} & a_{11} & \dots & a_{1n} \\
\vdots & \vdots && \vdots \\
a_{n0} & a_{n1} & \dots & a_{nn} 
\end{array}\right|
\ne 0.
$$
Ellers finnes enten ingen eller uendelig mange løsninger.

### Begrunnelse:

Gauss eliminasjonen vi gjør når vi løser systemet endrer ikke på om determinanten er lik null. Det vil si determinanten av matrisen på venstre siden av den vertikale streken vil enten være null i hvert steg, eller den vil være forskjellig fra null i hvert steg. 

Hvis vi til slutt får identitetsmatrisen $I$ på venstre siden av den vertikale streken da må den opprindelige matrisen ha determinant forskjellig fra null. 

Ellers vil vi til slutt ha en rad med bare nuller på venstre siden av den vertikale streken, og matriser med en slik null-rad har determinant null. 
Hvis det til høyre for denne raden står et tall som ikke er null finnes ingen løsning til ligningssystemet. 

Hvis det til høyre for alle rader på venstre siden av den vertikale streken som er blitt til null ved gauss eliminasjon står null, og det er minst en slik rad, da har denne raden ingen pivot posisjon, og derfor finnes frie parametre. Det vil si at systemet har uendelig mange løsninger. 

## Eksempel

Siden
$$ 
\left|\begin{array}{ccc}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{array}\right|
= 0
$$
Har lignngssystemet 

$$
\begin{array}{ccccccc}
    1x_0 &+& 2x_1& +& 3x_2& = &2 \\
    4x_0 &+& 5x_1& +& 6x_2& = &5 \\
    7x_0 &+& 8x_1& +& 9x_2& = &8 
\end{array}
$$


Enten ingen eller uendelig mange løsninger. Vi har allerede sett at det finnes en løsning for hvert valg av $x_2$.

## Oppgave

Finn alle $t$ slik at systemet

$$\begin{array}{ccccc}
(1-t)x & + & 2y & = & 0 \\
x & + & (2-t)y & = & 0
\end{array}$$

uendelig mange løsniger.

## Teorem

Gitt to $n \times n$-matriser $A$ og $B$ er $\det(A B) = \det(A) \det(B)$
