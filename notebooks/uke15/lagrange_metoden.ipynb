# Lagrange multiplikator metoden

Vi ser nå på to funksjoner $f, g \colon \mathbb{R}^n \to \mathbb{R}$.

Problemet vi stiller oss er å finne et minmum for $f$ på mengden
$$M = \{ \mathbf{x}\in \mathbb{R}^n \mid
g(\mathbf{x}) = 0 \}.$$ 

Vi antar at $f$ og $g$ er differensierbare og at
$\nabla g(\mathbf{x})
\ne 0$ for alle $\mathbf{x}\in M$.

I situasjonen over sier vi også at vi søker et minimum for $f$ under
betingelsen $g(x) = 0$.

## Teorem
*Hvis $\mathbf{p}$ er et minimumspunkt for $f$ under betingelsen
$g(\mathbf{x}) = 0$, da er vektorene $\nabla f(\mathbf{p})$ og
$\nabla g(\mathbf{p})$ parallelle. Det vi vil si at der finnes et tall
$\lambda$ slik at $\nabla f(\mathbf{p})
  = 
  \lambda \nabla g(\mathbf{p})$.*

[**Forklaring**](lagrange_forklaring.ipynb)

## Eksempel

La $f\colon \mathbb{R}^2 \to \mathbb{R}$ være funksjonen
$$f(\begin{bmatrix}x\\y\end{bmatrix}) = x^2 + 4y^2 -2x + 8y.$$ Vi vil
finne alle lokale minima for $f$ under betingelsen $x + 2y = 7$.

La $g\colon \mathbb{R}^2 \to \mathbb{R}$ være funksjonen
$g(\begin{bmatrix}x\\y\end{bmatrix})
  = x + 2y - 7$.

### Oppgave

Finn formler for $\nabla g(\begin{bmatrix} x\\y \end{bmatrix})$ og $\nabla f(\begin{bmatrix} x\\y \end{bmatrix})$.

Geometrisk sett sier teoremet til Lagrange at tangentene til nivåkurvene
for $f$ og $g$ må være parallelle. (Egentlig sier det at gradiengene,
som står vinkelrett på tangentene, må være parallelle, men det er det
samme som å si at tangentene er parallelle.)

### Oppgave

Finn $x, y$ og $\lambda$ slik at $\nabla f(\begin{bmatrix} x\\y \end{bmatrix}) = \lambda \nabla g(\begin{bmatrix} x\\y \end{bmatrix})$ og $g(\begin{bmatrix} x\\y \end{bmatrix}) = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# @title Plotter
# import matplotlib.pyplot as plt

# from matplotlib import cm
# from mpl_toolkits.mplot3d import axes3d

# Define the function
def f(x):
    x1, x2 = x
    return x1**2 + 4*x2**2 - 2*x1 + 8*x2
# Create a grid of x and y values
x1 = np.linspace(-8, 8, 100)
x2 = np.linspace(-7, 9/2, 100)
x = np.meshgrid(x1, x2)

# Compute the function values
x3 = f(x)

# Create a 3D plot
fig, ax = plt.subplots()
# CS = ax.contour(X, Y, Z)
ax.imshow(x3[::-1], extent=[np.min(x1), np.max(x1), np.min(x2), np.max(x2)])
#ax.scatter(x[0], x[1], c=x3, cmap='viridis', edgecolor='none', alpha=0.6)
ax.contour(x1, x2, x3, colors='red', alpha=.4) #, cmap=cm.coolwarm)  # Plot contour curves
ax.plot([-2, 8], [9/2, -1/2], linewidth=5, color='black', label=r'g=0')
plt.legend()
plt.title(r'f represented by color')
ax.set_aspect(1)
```

## Eksempel

Vi bytter om på rollene til $f$ og $g$:

La $g\colon \mathbb{R}^2 \to \mathbb{R}$ være funksjonen
$$g(\begin{bmatrix}x\\y\end{bmatrix}) = x^2 + 4y^2 -2x + 8y -27$$ og la
$$f(\begin{bmatrix}x\\y\end{bmatrix}) = x + 2y.$$ Vi vil minimere $f$
under betingelsen $$g(\begin{bmatrix}x\\y\end{bmatrix}) = 0.$$ Vi vil
finne alle lokale minima for $f$ under betingelsen
$g(\begin{bmatrix}x\\y\end{bmatrix}) = 0$.

```python
# @title Plotter
# import matplotlib.pyplot as plt

# from matplotlib import cm
# from mpl_toolkits.mplot3d import axes3d

# Define the function
def f(x):
    x1, x2 = x
    return x1 + 2*x2
    # return x1**2 + 4*x2**2 - 2*x1 + 8*x2
# Create a grid of x and y values
x1 = np.linspace(-8, 8, 100)
x2 = np.linspace(-7, 9/2, 100)
x = np.meshgrid(x1, x2)
t = np.linspace(0, 2*np.pi, num=100)


# Compute the function values
x3 = f(x)

# Create a 3D plot
fig, ax = plt.subplots()
# CS = ax.contour(X, Y, Z)
ax.plot(np.cos(t)*np.sqrt(32) + 1, np.sin(t)*np.sqrt(8) - 1, color='black', label=r'g=0')
ax.imshow(x3[::-1], extent=[np.min(x1), np.max(x1), np.min(x2), np.max(x2)])
ax.scatter([-3, 5],[-3, 1], color='black')
#ax.scatter(x[0], x[1], c=x3, cmap='viridis', edgecolor='none', alpha=0.6)
#ax.contour(x1, x2, x3, colors='red', alpha=.4) #, cmap=cm.coolwarm)  # Plot contour curves
#ax.plot([-2, 8], [9/2, -1/2], linewidth=5, color='black')
plt.legend()
plt.title(r'f represented by color')
ax.set_aspect(1)
```

Som før får vi de to
likningene $$2x - \lambda = 2$$ og $$8y - 2\lambda = -8.$$ Begingelsen
$g(x) = 0$ blir nå $$x^2 + 4y^2 -2x + 8y -27 = 0$$

Jeg har lyst at dette skal bli til en ligning av grad to i en ukjent.
Derfor skriver jeg ved hjelp av de første to likningene
$$x = \lambda/2 + 1$$ og $$y = \lambda/4 - 1.$$ Dette setter jeg inn i
den andre ligningen:

$$\begin{aligned}
    0 =
    &x^2 + 4y^2 -2x + 8y -27 \\
    =&
    (\lambda/2 + 1)^2 + 4(\lambda/4 - 1)^2 -2(\lambda/2 + 1) + 8(\lambda/4 - 1) -27\\
    =&
    \lambda^2/4 + \lambda + 1 + \lambda^2/4 - 2\lambda + 4 -\lambda - 2 + 2\lambda - 8 -27\\
    =&
    \lambda^2/2 - 32.
\end{aligned}$$
Her står $\lambda^2 = 64 = 8^2$. Derfor er
$\lambda = \pm 8$.

Setter jeg inn $\lambda = 8$ får jeg at $x = 5$ og $y = 1$. Setter jeg
inn $\lambda = -8$ får jeg at $x = -3$ og $y = -3$.

Siden
$$f(\begin{bmatrix}5\\1\end{bmatrix}) = 5^2 + 4 - 2\cdot 5 + 8 = 27$$ og
$$f(\begin{bmatrix}-3\\-3\end{bmatrix}) = 9 + 12 + 6 - 24 = 3$$ må
$f(\begin{bmatrix}-3\\-3\end{bmatrix})$ være et minimum for $f$ under
begingelsen $g = 0$. Altså er punktet
$\mathbf{p} = \begin{bmatrix}-3\\-3\end{bmatrix}$ et minimum for $f$
under betingelsen $g(\mathbf{p}) = 0$.
