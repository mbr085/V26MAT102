## Teorem
*Hvis $\mathbf{p}$ er et minimumspunkt for $f$ under betingelsen
$g(\mathbf{x}) = 0$, da er vektorene $\nabla f(\mathbf{p})$ og
$\nabla g(\mathbf{p})$ parallelle. Det vi vil si at der finnes et tall
$\widetilde \lambda$ slik at $\nabla f(\mathbf{p})
  = \widetilde
  \lambda \nabla g(\mathbf{p})$.*

**Forklaring** (Kan ignoreres)

Vi følger nøye strategien fra gradient descent: La
$\mathbf{y}\colon \mathbb{R}\to \mathbb{R}^n$ være en løsning til
differensialligningen
$$\mathbf{y}'(t) = - \nabla f(\mathbf{y}(t)) + \lambda(t) \nabla g(\mathbf{y}(t))$$
der $\lambda \colon \mathbb{R}\to \mathbb{R}$ er funksjonen
$$\lambda(t) = \frac{\nabla f(\mathbf{y}(t)) \cdot \nabla g(\mathbf{y}(t))}{\nabla g(\mathbf{y}(t)) \cdot \nabla g(\mathbf{y}(t))}.$$

Fra kjerneregelen har vi at
$(g \circ \mathbf{y})'(t) = \nabla g(\mathbf{y}(t)) \cdot \mathbf{y}'(t).$

### Oppgave 
Vis at $(g \circ y)'(t) = 0$ for alle $t$.

---

Siden
$|\nabla f(\mathbf{y}(t)) \cdot \nabla g(\mathbf{y}(t)) \le |\nabla f(\mathbf{y}(t)| |\nabla g(\mathbf{y}(t))||$,
ser vi at

$$\begin{aligned}
  (f \circ \mathbf{y})'(t)
  &= \nabla f(\mathbf{y}(t)) \cdot \mathbf{y}'(t)\
  &= -|\nabla f(\mathbf{y}(t)|^2 +
  \frac{(\nabla f(\mathbf{y}(t)) \cdot \nabla g(\mathbf{y}(t)))^2}
  {(\nabla g(\mathbf{y}(t)) \cdot \nabla g(\mathbf{y}(t)))}
  \le\
  &= -|\nabla f(\mathbf{y}(t)|^2 +
  \frac{|\nabla f(\mathbf{y}(t))|^2 |\nabla g(\mathbf{y}(t))|^2}
  {|\nabla g(\mathbf{y}(t))|^2} = 0.
\end{aligned}$$

Som for gradient descent ser vi at dersom $\mathbf{y}(t)$ går mot et
punkt $\mathbf{p}$ for $t \to \infty$, da har vi at
$(f \circ \mathbf{y})'(t) \to 0$ for $t \to \infty$.

La oss skrive
$$\widetilde \lambda = \lim_{t \to \infty} \lambda(t) = \frac{\nabla f(\mathbf{p}) \cdot \nabla g(\mathbf{p})}{\nabla g(\mathbf{p}) \cdot \nabla g(\mathbf{p})}.$$

Siden
$$(f \circ \mathbf{y})'(t) = -| \nabla f(\mathbf{y}(t))|^2 + \lambda(t) \nabla f(\mathbf{y}(t)) \cdot \nabla
g(\mathbf{y}(t))$$

for alle $t$, ser vi ved å la $t$ gå mot uendelig at

$$-| \nabla f(\mathbf{p})|^2 + \widetilde \lambda \nabla f(\mathbf{p}) \cdot \nabla g(\mathbf{p}) =0.$$

Setter vi inn uttrykket for $\widetilde \lambda$, og ganger vi med nevneren i dette uttrykket får vi

$$-|\nabla g(\mathbf{p})|^2 | \nabla f(\mathbf{p})|^2 + (\nabla f(\mathbf{p}) \cdot \nabla g(\mathbf{p}))^2 =0.$$

Dette bare
er mulig dersom $\nabla f(\mathbf{p})$ og $\nabla g(\mathbf{p})$ er parallelle, og mere presist må vi ha
$\nabla f(\mathbf{p}) = \widetilde \lambda \nabla g(\mathbf{p})$.

Omvendt, hvis $\mathbf{p}$ er et minimum for $f$ under betingelsen
$g(\mathbf{x}) = 0$, da må vi ha

$$\nabla f(\mathbf{p}) = \widetilde \lambda \nabla g(\mathbf{p}) \quad\text{for en
$\widetilde \lambda.$}$$

Grunnen til dette er at ellers vil differensialligningen over ha $\mathbf(f \circ \mathbf{y})'(0) < 0$ for løsningen $\mathbf{y}(t)$ med $\mathbf{y}(0) = \mathbf{p}$. Det vil si at det finnes $t$ villkårlig nær $0$ med $\mathbf(f \circ \mathbf{y})(t) < \mathbf(f \circ \mathbf{y})(0)$. Derfor kan $\mathbf{p} = \mathbf{y}(0)$
ikke være et lokalt minimumspunkt for $f$ under betingelsen $g(\mathbf{p}) = 0$.
