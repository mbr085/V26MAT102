## Oppgave

La $f\colon \mathbb{R}^2 \to \mathbb{R}$ være funksjonen
$$f(\begin{bmatrix}x\\y\end{bmatrix}) = x^2 + 4y^2 -2x + 8y.$$ Vi vil
finne alle lokale minima for $f$ under betingelsen $x + 2y = 7$.

La $g\colon \mathbb{R}^2 \to \mathbb{R}$ være funksjonen
$g(\begin{bmatrix}x\\y\end{bmatrix})
  = x + 2y - 7$.

Finn $x, y$ og $\lambda$ slik at $\nabla f(\begin{bmatrix} x\\y \end{bmatrix}) = \lambda \nabla g(\begin{bmatrix} x\\y \end{bmatrix})$ og $g(\begin{bmatrix} x\\y \end{bmatrix}) = 0$

## Løsningsforslag

Gradientene til $f$ og $g$ er

$$\nabla f(\begin{bmatrix} x\\y \end{bmatrix}) = 
\begin{bmatrix}
2x - 2 \\
8y + 8
\end{bmatrix}$$
og
$$\nabla g(\begin{bmatrix} x\\y \end{bmatrix}) = 
\begin{bmatrix}
1 \\
2
\end{bmatrix}$$

Det gir
$$\nabla f(\begin{bmatrix} x\\y \end{bmatrix}) - \lambda \nabla g(\begin{bmatrix} x\\y \end{bmatrix}) = 
\begin{bmatrix}
2 & 0 & -1 \\
0 & 8 & -2
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
\lambda 
\end{bmatrix}
+
\begin{bmatrix}
-2 \\
8
\end{bmatrix}
$$

Legger vi til en rad med ligningen $0 = x + 2y - 7$ får vi matrisen

$$
0 = 
\begin{bmatrix}
2 & 0 & -1 \\
0 & 8 & -2 \\
1 & 2 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
\lambda 
\end{bmatrix}
+
\begin{bmatrix}
-2 \\
8\\
-7
\end{bmatrix}
$$


Dette kan skrives
$$
\begin{bmatrix}
2 & 0 & -1 \\
0 & 8 & -2 \\
1 & 2 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
\lambda 
\end{bmatrix}
 = 
\begin{bmatrix}
2 \\
-8\\
7
\end{bmatrix}
$$


Dette kan vi løse ved matrisereduksjon.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    x1, x2 = x
    return x1**2 + 4*x2**2 - 2*x1 + 8*x2
    
```

```python
A = np.array([
    [2., 0, -1, 2],
    [0, 8, -2, -8],
    [1, 2, 0, 7]
])
```
