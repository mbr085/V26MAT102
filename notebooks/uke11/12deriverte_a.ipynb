# Derivasjon

## Vektorer og totalderiverte

```python
# @title Import
import matplotlib.pyplot as plt
import numpy as np
```

**Definisjon 1**. Funksjonen $\mathbf{F}$ er *deriverbar* i punktet
$\mathbf{x}_0$ hvis det finnes en $m \times n$-matrise
$\mathbf{F}'(\mathbf{x}_0)$ slik at
$$\lim_{\mathbf{h}\to 0} \frac{|\mathbf{F}(\mathbf{x}_0+\mathbf{h}) - \mathbf{F}(\mathbf{x}_0) -
  \mathbf{F}'(\mathbf{x}_0) \cdot \mathbf{h}|}{|\mathbf{h}|} = 0.$$ Her
tas grensen over $\mathbf{h} \in \mathbb{R}^n$ og $| \mathbf{h}|$ er
lengden til vektoren $\mathbf{h}$.

Funksjonen $\mathbf{F}$ er *derivervar* hvis den er deriverbar i alle
punktene i $\mathbb{R}^n$.

Matrisen $\mathbf{F}(\mathbf{x}_0)'$ kalles den deriverte av
$\mathbf{F}$ i punktet $\mathbf{x}_0$. Et annet navn for denne matrisen
er *Jacobimatrisen* til $\mathbf{F}$ i punktet $\mathbf{x}_0$.

Vi går ikke inn på detalj om grenseverdier i flere variabler. 

Alt vi
trenger å vite er at de har samme regneregler som grenseverdier i en
variabel og at grenseverdier kan beregnes komponentvis. 

En konsekvens er
at de generelle derivasjonsreglene for funksjoner av en variabel også
gjelder for vektorfunksjoner. 

En annen konsekvens er at
$$\mathbf{F}'(\mathbf{x}) = \begin{bmatrix} F'_1(\mathbf{x}) \\ F'_2(\mathbf{x})
\\ \vdots \\ F'_m(\mathbf{x}) \end{bmatrix},$$ hvor matrisen på høyre
side er en $m \times n$-matrise fremkommet ved å stable $m$ matriser av
formen $(1 \times n)$ på hverandre. Det vil si $m$ radvektorer, hvor
hver radvektor er den deriverte av en funksjon $F_i$.

**Eksempel 2**. 

For 
$$\mathbf F(\begin{bmatrix} x\\y\end{bmatrix}) = \begin{bmatrix} x^2 + y^2 \\ x^2 - y^2\end{bmatrix}$$
er
$$F_1(\begin{bmatrix} x\\y\end{bmatrix}) = x^2 + y^2$$
og 
$$F_2(\begin{bmatrix} x\\y\end{bmatrix}) = x^2 - y^2.$$

Den deriverte til $F_1$ er
$$F_1'(\begin{bmatrix} x\\y\end{bmatrix})= \begin{bmatrix} 2x & 2y \end{bmatrix}.$$
Her er $2x$ den deriverte til $F_1$ som funksjon av $x$ og 
$2y$ den deriverte til $F_1$ som funksjon av $y$. 

Den deriverte til $F_2$ er
$$F_2'(\begin{bmatrix} x\\y\end{bmatrix})= \begin{bmatrix} 2x & -2y \end{bmatrix}.$$

Den deriverte til $\mathbf F$ er 
$$\mathbf F'(\begin{bmatrix} x\\y\end{bmatrix}) = \begin{bmatrix} 2x & 2y \\ 2x & -2y\end{bmatrix}.$$

**Eksempel 3**. 

La $A$ være $2 \times 2$-matrisen
$$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}.$$ Da er funksjonen
$\mathbf{F}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved
$$\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}= \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
  \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} x + 2y \\ 3x
  + 4y \end{bmatrix}$$
en vektorfunksjon.

Her skriver vi $\mathbf{x}$ for et punkt
$\mathbf{x}= \begin{bmatrix} x \\ y \end{bmatrix}$ i $\mathbb{R}^2$.
Siden et punkt i $\mathbb{R}^2$ er det samme som en vektor, gir det her mening å skrive
skrive $\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}$. 

Gitt et punkt
$\mathbf{b} = \begin{bmatrix}b_1\\b_2\end{bmatrix}$  i
$\mathbb{R}^2$
kan vi konstruere en vektorfunsjon
$\mathbf{G}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved

$$\mathbf{G}(\mathbf{x}) = A \cdot \mathbf{x}+ \mathbf{b} = \begin{bmatrix} x_1 + 2x_2 \\ 3x_1 + 4x_2
    \end{bmatrix} + \begin{bmatrix} b_1 \\ b_2 \end{bmatrix} = \begin{bmatrix}
  x_1 + 2x_2 + b_1 \\ 3x_1 + 4x_2 + b_2 \end{bmatrix}.$$

Den deriverte 
$\mathbf{G}'(\mathbf{x})$ er en matrise slik at
$$\lim_{\mathbf{h}\to 0} \frac{|\mathbf{G}(\mathbf{x}+\mathbf{h}) - \mathbf{G}(\mathbf{x}) -
  \mathbf{G}'(\mathbf{x}) \cdot \mathbf{h}|}{|\mathbf{h}|} = 0.$$
Regner vi på dette får vi:
$$\begin{align*}
    &
    \lim_{\mathbf{h}\to 0} \frac{|\mathbf{G}(\mathbf{x}+\mathbf{h}) -
    \mathbf{G}(\mathbf{x}) -
    \mathbf{G}'(\mathbf{x}) \cdot \mathbf{h}|}{|\mathbf{h}|} \\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{|\left(A \cdot (\mathbf{x}+\mathbf{h})  +
    \mathbf{b}\right) - \left(A \cdot \mathbf{x}  + \mathbf{b}\right) -
    \mathbf{G}'(\mathbf{x}) \cdot \mathbf{h}|}{|\mathbf{h}|}\\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{|A \cdot (\mathbf{x}+\mathbf{h})  -
    A \cdot \mathbf{x}  -
    \mathbf{G}'(\mathbf{x}) \cdot \mathbf{h}|}{|\mathbf{h}|} \\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{|A \cdot \mathbf{h}  -
    \mathbf{G}'(\mathbf{x}) \cdot \mathbf{h}|}{|\mathbf{h}|}.
\end{align*}$$
Setter vi $\mathbf{G}'(\mathbf{x}) = A$ blir dette null. 
Derfor er $\mathbf{G}'(\mathbf{x}) = A$ den
deriverte til $\mathbf{G}$ i punktet $\mathbf{x}$.

**Definisjon 4**. *Lineærtilnærmingen* til en vektorfunksjon
$\mathbf{F}$ som er deriverbar i punktet $\mathbf{x}_0$ er gitt ved
$$L(\mathbf{x}) = \mathbf{F}(\mathbf{x}_0) + \mathbf{F}'(\mathbf{x}_0) \cdot (\mathbf{x} - \mathbf{x}_0).$$

**Eksempel 5**. La igjen $\mathbf{G}$ være vektorfunksjonen fra eksempel
3. Lineærtilnærmingen til
$\mathbf{G}$ i punktet $\mathbf{x}_0$ er gitt ved
$$L(\mathbf{x}) = \mathbf{G}(\mathbf{x}_0) + \mathbf{G}'(\mathbf{x}_0) \cdot (\mathbf{x}- \mathbf{x}_0) = A \cdot \mathbf{x}_0 + \mathbf{b} + A \cdot (\mathbf{x}- \mathbf{x}_0).$$
Siden
$$A \cdot \mathbf{x}_0 + \mathbf{b} + A \cdot (\mathbf{x}- \mathbf{x}_0) = A \cdot \mathbf{x}_0 + A
  \cdot \mathbf{x}- A \cdot \mathbf{x}_0 + \mathbf{b} = A \cdot \mathbf{x}+ \mathbf{b},$$
ser vi at $L(\mathbf{x}) = \mathbf{G}(\mathbf{x})$.

**Eksempel 6**. La $f \colon \mathbb{R}\to \mathbb{R}$ være en vanlig
funksjon. Betingelsen for at $f$ er deriverbar i $x_0$ kan skrives:
$$\lim_{h\to 0} \frac{f(x_0+h) - f(x_0) - f'(x_0) \cdot h}{h} =
  \lim_{h\to 0} \frac{f(x_0+h) - f(x_0)}{h} - f'(x_0) = 0.$$ Her står
altså $$f'(x_0) = \lim_{h\to 0} \frac{f(x_0+h) - f(x_0)}{h}.$$ Dette er
den vanlige definisjonen av den deriverte av en funksjon.

Lineærtilnærmingen til $f$ i punktet $x_0$ er gitt ved
$$L(x) = f(x_0) + f'(x_0) \cdot (x - x_0).$$ Grafen til $L(x)$ er
tangenten i punktet $x_0$ til grafen for $f$.

```python
# @title Tangent
# Definer funksjonen
def f(x):
  return (x - 1) * (x - 1) * (x - 2)  # Funksjonen f(x) = (x-1)²(x-2)

# Definer den deriverte av funksjonen
def df(x):
  return 2 * (x - 1) * (x - 2) + (x - 1) * (x - 1)  # Den deriverte av f(x)

# Definer intervallet
x = np.linspace(0, 3, 400)  # Lag 400 jevnt fordelte punkter mellom 0 og 3

# Beregn funksjonsverdiene
y = f(x)  # Beregn f(x) for alle punktene i x

# Punkt for tangenten
x0 = 2  # x-verdien der vi vil tegne tangenten
y0 = f(x0)  # y-verdien til tangeringspunktet
m = df(x0)  # Stigningstallet til tangenten

# Beregn tangentlinjen
x_tan = np.linspace(1., 3., 100)  # Intervall for tangentlinjen
y_tan = m * (x_tan - x0) + y0  # Beregn y-verdiene til tangentlinjen

# Lag plottet
plt.plot(x, y, label=r'$f(x) = (x-1)^2(x-2)$')  # Plott funksjonen
plt.plot(x_tan, y_tan, label='Tangent ved x=2')  # Plott tangentlinjen
plt.scatter(x0, y0, color='red', label='Tangeringspunkt')  # Marker tangeringspunktet

# Tilpass plottet
plt.xlabel('x')  # Navn på x-aksen
plt.ylabel('f(x)')  # Navn på y-aksen
plt.title('Plott av f(x) og dens tangent ved x=2')  # Tittel på plottet
plt.legend()  # Vis forklaring
plt.grid(True)  # Vis rutenett
ax = plt.gca()
ax.set_aspect(1)

# Vis plottet
plt.show()
```

**Eksempel 7**. La $\mathbf{y} \colon \mathbb{R}\to \mathbb{R}^2$ være
en kurve i planet. Betingelsen for at $\mathbf{y}$ er deriverbar i $x_0$
kan skrives:
$$\lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0) - \mathbf{y}'(x_0) \cdot h}{h} =
  \lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0)}{h} - \mathbf{y}'(x_0) = 0.$$
Her står altså
$$\mathbf{y}'(x_0) = \lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0)}{h}.$$
Som bemerket over kan denne grensen beregens komponentvis slik at hvis
$$\mathbf{y}(t) =
  \begin{bmatrix} y_1(t) \\ y_2(t) \end{bmatrix},$$ så er
$$\mathbf{y}'(t) =
  \begin{bmatrix} y_1'(t) \\ y_2'(t) \end{bmatrix}.$$ Lineærtilnærmingen
til $\mathbf{y}$ i punktet $t_0$ er gitt ved
$$L(t) = \mathbf{y}(t_0) + \mathbf{y}'(t_0) \cdot (t - t_0).$$ Dette er
en parametrisering av tangenten til kurven i punktet $\mathbf{y}(t_0)$.

**Oppgave 2**

La $\mathbf{y} \colon \mathbb{R}\to \mathbb{R}^2$ være gitt ved 
$$\mathbf{y}(t) =
\begin{bmatrix}
\cos(t) \\

\end{bmatrix}$$
Beregn lineærtilnærmingen til $\mathbf{y}$ i $t_0 = 0$.

```python
# @title Tangent
# Definer funksjonen
def f(x):
  return (x - 1) * (x - 1) * (x - 2)  # Funksjonen f(x) = (x-1)²(x-2)

# Definer den deriverte av funksjonen
def df(x):
  return 2 * (x - 1) * (x - 2) + (x - 1) * (x - 1)  # Den deriverte av f(x)

# Definer intervallet
x = np.linspace(0, 3, 400)  # Lag 400 jevnt fordelte punkter mellom 0 og 3

# Beregn funksjonsverdiene
y = f(x)  # Beregn f(x) for alle punktene i x

# Punkt for tangenten
x0 = 2  # x-verdien der vi vil tegne tangenten
y0 = f(x0)  # y-verdien til tangeringspunktet
m = df(x0)  # Stigningstallet til tangenten

# Beregn tangentlinjen
x_tan = np.linspace(1., 3, 100)  # Intervall for tangentlinjen
y_tan = m * (x_tan - x0) + y0  # Beregn y-verdiene til tangentlinjen

# Lag plottet
plt.plot(x, y, label=r'$y(t) = (t, \ (t-1)^2(t-2))$')  # Plott funksjonen
plt.plot(x_tan, y_tan, label='Tangent ved x=2')  # Plott tangentlinjen
plt.scatter(x0, y0, color='red', label='Tangeringspunkt')  # Marker tangeringspunktet

# Tilpass plottet
plt.xlabel('x')  # Navn på x-aksen
plt.ylabel('f(x)')  # Navn på y-aksen
plt.title('Plott av y(t) og dens tangent ved t=2')  # Tittel på plottet
plt.legend()  # Vis forklaring
plt.grid(True)  # Vis rutenett
ax = plt.gca()
ax.set_aspect(1)


# Vis plottet
plt.show()
```

```python
# @title Tangent
def yy(t):
    return np.vstack([2*np.cos(t), np.sin(t)])

def dyy(t):
    return np.vstack([-2*np.sin(t), np.cos(t)])

# Definer intervallet
t = np.linspace(0, 2*np.pi, 400)  # Lag 400 jevnt fordelte punkter mellom 0 og 3

# Beregn funksjonsverdiene
y = yy(t)  # Beregn f(x) for alle punktene i x

# Punkt for tangenten
t0 = np.pi / 4

# Beregn tangentlinjen
t_tan = np.linspace(-1, +1, 100)  # Intervall for tangentlinjen
tan = yy(t0) + t_tan * dyy(t0)
# Lag plottet
plt.plot(*yy(t), label=r'$y(t) = (2\cos(t), \sin(t))$')  # Plott funksjonen
plt.plot(*tan, label='Tangent ved x=2')  # Plott tangentlinjen
plt.scatter(*yy(t0), color='red', label='Tangeringspunkt')  # Marker tangeringspunktet

# Tilpass plottet
plt.xlabel('x')  # Navn på x-aksen
plt.ylabel('f(x)')  # Navn på y-aksen
plt.title('Plott av y(t) og dens tangent ved t=2')  # Tittel på plottet
ax = plt.gca()
ax.set_aspect(1)
plt.legend()  # Vis forklaring
plt.grid(True)  # Vis rutenett
```
