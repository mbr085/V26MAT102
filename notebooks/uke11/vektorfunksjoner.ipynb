# Vektorfunksjoner

En *vektorfunksjon* er en funksjon som tar en verdi i $\mathbb{R}^n$ og
gir en vektor i $\mathbb{R}^m$. 

En vektorfunksjon kan skrives som en
funksjon 
$$\mathbf{F}:\mathbb{R}^n\to\mathbb{R}^m$$

som tar en vektor
$\mathbf{x}\in\mathbb{R}^n$ og gir en vektor
$\mathbf{F}(\mathbf{x})\in\mathbb{R}^m$. 

En vektorfunksjon kan også
skrives som en liste av $m$ funksjoner av $n$ variabler, for eksempel
$$\mathbf{F}(\mathbf{x}) = \begin{bmatrix} F_1(\mathbf{x}) \\ F_2(\mathbf{x}) \\ \vdots \\ F_m(\mathbf{x}) \end{bmatrix}.$$
I teksten under kommer jeg ofte til å skrive "funksjon" når jeg mener
"vektorfunksjon".

**Eksempel 1**. La $A$ være $2 \times 2$-matrisen
$$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}.$$ Da er funksjonen
$\mathbf{F}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved
$$\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}= \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
  \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} x_1 + 2x_2 \\ 3x_1
  + 4x_2 \end{bmatrix}.$$ en vektorfunksjon.

Her skriver vi $\mathbf{x}$ for et punkt
$\mathbf{x}= \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ i $\mathbb{R}^2$.
Siden et punkt i $\mathbb{R}^2$ er det samme som en vektor, kan vi
skrive $\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}$. Det er vanskelig å
tegne grafen til denne funksjonen fordi den lever i fire dimensjoner.
Det vi kan gjøre er å velge noen vektorer og se hvorhen de blir
avbildet. For eksempel er
$$\mathbf{F}\left(\begin{bmatrix} 1 \\ 0 \end{bmatrix}\right) = \begin{bmatrix} 1 \\ 3 \end{bmatrix}.$$

```python
# @title Import
import matplotlib.pyplot as plt
import numpy as np
```

```python
# @title Plotter
# Definer matrisen A
A = np.array([[1, 2], [3, 4]])

# Definer vektorene
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([1, 1])

# Anvend transformasjonen
Av1 = A @ v1  # Av1 er resultatet av å multiplisere A med v1
Av2 = A @ v2  # Av2 er resultatet av å multiplisere A med v2
Av3 = A @ v3  # Av3 er resultatet av å multiplisere A med v3

# Lag plottet
plt.figure(figsize=(8, 8))  # Sett størrelsen på figuren

# Plott de originale vektorene
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')  # Plott v1 i blått
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')  # Plott v2 i rødt
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', label='v3')  # Plott v3 i grønt

# Plott de transformerte vektorene (bildene)
plt.quiver(0, 0, Av1[0], Av1[1], angles='xy', scale_units='xy', scale=1, color='blue', linestyle='dashed')  # Plott Av1 med stiplet linje
plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=1, color='red', linestyle='dashed')  # Plott Av2 med stiplet linje
plt.quiver(0, 0, Av3[0], Av3[1], angles='xy', scale_units='xy', scale=1, color='green', linestyle='dashed')  # Plott Av3 med stiplet linje

# Legg til navn på de originale vektorene
plt.annotate(r"$\mathbf{x}$", v1 + 0.1, color='blue')  # Legg til navn på v1, litt forskjøvet for bedre synlighet
plt.annotate(r"$\mathbf{y}$", v2 + 0.1, color='red')  # Legg til navn på v2, litt forskjøvet for bedre synlighet
plt.annotate(r"$\mathbf{z}$", v3 + 0.1, color='green')  # Legg til navn på v3, litt forskjøvet for bedre synlighet

# Legg til navn på de transformerte vektorene
plt.annotate(r"$\mathbf{F}(\mathbf{x})$", Av1 + [-0.1, 0.1], color='blue')  # Legg til navn på Av1, litt forskjøvet
plt.annotate(r"$\mathbf{F}(\mathbf{y})$", Av2 + 0.1, color='red')  # Legg til navn på Av2, litt forskjøvet
plt.annotate(r"$\mathbf{F}(\mathbf{z})$", Av3 + 0.1, color='green')  # Legg til navn på Av3, litt forskjøvet

matrix_text = u"⎡1 2⎤\n⎣3 4⎦"
plt.text(5, 8, matrix_text, fontsize=12)
plt.text(4.4, 8.2, 'A =', fontsize=12)
plt.text(1,8.2, 'F(x) = Ax', fontsize=12)

# Legg til aksenavn og tittel
plt.xlabel("$x_1$")  # Navn på x-aksen
plt.ylabel("$x_2$")  # Navn på y-aksen
plt.xlim(-1, 7)  # Sett grenser for x-aksen
plt.ylim(-1, 9)  # Sett grenser for y-aksen
plt.title("Transformasjon av vektorer ved matrise A")  # Tittel på plottet
plt.grid(True)  # Vis rutenett
#plt.legend()  # Vis forklaring (kan aktiveres om ønskelig)
ax = plt.gca()
ax.set_aspect(1)

# Vis plottet
plt.show()
```

**Eksempel 2**. Vi fortsetter eksempel
<a href="#lineærfunksjon" data-reference-type="ref"
data-reference="lineærfunksjon">1</a> og lar
$\mathbf{b} = \begin{bmatrix}b_1\\b_2\end{bmatrix}$ være et punkt i
$\mathbb{R}^2$. 

Da kan vi konstruere en vektorfunsjon
$\mathbf{G}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved

$$\mathbf{G}(\mathbf{x}) = A \cdot \mathbf{x}+ \mathbf{b} = \begin{bmatrix} x_1 + 2x_2 \\ 3x_1 + 4x_2
    \end{bmatrix} + \begin{bmatrix} b_1 \\ b_2 \end{bmatrix} = \begin{bmatrix}
  x_1 + 2x_2 + b_1 \\ 3x_1 + 4x_2 + b_2 \end{bmatrix}.$$

```python
# @title Plotter
# Definer matrisen A
A = np.array([[1, 2], [3, 4]])

# definer vektoren b
b = np.array([2, -1])

# Definer vektorene
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([1, 1])

# Anvend transformasjonen
Gv1 = A @ v1 + b # Gv1 er resultatet av å multiplisere A med v1
Gv2 = A @ v2 + b # Gv2 er resultatet av å multiplisere A med v2
Gv3 = A @ v3 + b # Gv3 er resultatet av å multiplisere A med v3

# Lag plottet
plt.figure(figsize=(8, 8))  # Sett størrelsen på figuren

# Plott de originale vektorene
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')  # Plott v1 i blått
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')  # Plott v2 i rødt
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', label='v3')  # Plott v3 i grønt

# Plott de transformerte vektorene (bildene)
plt.quiver(0, 0, Gv1[0], Gv1[1], angles='xy', scale_units='xy', scale=1, color='blue', linestyle='dashed')  # Plott Gv1 med stiplet linje
plt.quiver(0, 0, Gv2[0], Gv2[1], angles='xy', scale_units='xy', scale=1, color='red', linestyle='dashed')  # Plott Gv2 med stiplet linje
plt.quiver(0, 0, Gv3[0], Gv3[1], angles='xy', scale_units='xy', scale=1, color='green', linestyle='dashed')  # Plott Gv3 med stiplet linje

# Legg til navn på de originale vektorene
plt.annotate(r"$\mathbf{x}$", v1 + 0.1, color='blue')  # Legg til navn på v1, litt forskjøvet for bedre synlighet
plt.annotate(r"$\mathbf{y}$", v2 + 0.1, color='red')  # Legg til navn på v2, litt forskjøvet for bedre synlighet
plt.annotate(r"$\mathbf{z}$", v3 + 0.1, color='green')  # Legg til navn på v3, litt forskjøvet for bedre synlighet

# Legg til navn på de transformerte vektorene
plt.annotate(r"$\mathbf{F}(\mathbf{x})$", Gv1 + [-0.1, 0.1], color='blue')  # Legg til navn på Gv1, litt forskjøvet
plt.annotate(r"$\mathbf{F}(\mathbf{y})$", Gv2 + 0.1, color='red')  # Legg til navn på Gv2, litt forskjøvet
plt.annotate(r"$\mathbf{F}(\mathbf{z})$", Gv3 + 0.1, color='green')  # Legg til navn på Gv3, litt forskjøvet

matrix_text = u"⎡1 2⎤\n⎣3 4⎦"
plt.text(5, 8, matrix_text, fontsize=12)
plt.text(4.4, 8.2, 'A =', fontsize=12)
vector_text = u"⎡ 2⎤\n⎣-1⎦"
plt.text(4.4, 7.2, 'b =', fontsize=12)
plt.text(5, 7, vector_text, fontsize=12)
plt.text(1,7.5, 'G(x) = Ax + b', fontsize=12)
# Legg til aksenavn og tittel
plt.xlabel("$x_1$")  # Navn på x-aksen
plt.ylabel("$x_2$")  # Navn på y-aksen
plt.xlim(-1, 7)  # Sett grenser for x-aksen
plt.ylim(-1, 9)  # Sett grenser for y-aksen
plt.title("Transformasjon av vektorer ved matrise A")  # Tittel på plottet
plt.grid(True)  # Vis rutenett
#plt.legend()  # Vis forklaring (kan aktiveres om ønskelig)
ax = plt.gca()
ax.set_aspect(1)

# Vis plottet
plt.show()
```
